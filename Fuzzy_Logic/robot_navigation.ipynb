```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
sleft = ctrl.Antecedent(np.arange(0,2.1,0.1),'sleft') 
sright = ctrl.Antecedent(np.arange(0,2.1,0.1),'sright')
scen = ctrl.Antecedent(np.arange(0,2.1,0.1),'scen')
mleft = ctrl.Consequent(np.arange(-100,101,1),'mleft')
mright = ctrl.Consequent(np.arange(-100,101,1),'mright')
```

```python
sleft['short'] = fuzz.trimf(sleft.universe,[0,0,0.8])
sleft['medium'] = fuzz.trimf(sleft.universe,[0.2,1,1.7])
sleft['long'] = fuzz.trimf(sleft.universe,[1.4,2,2])
sleft.view()
```

```python
scen['short'] = fuzz.trimf(scen.universe,[0,0,0.8])
scen['medium'] = fuzz.trimf(scen.universe,[0.2,1,1.7])
scen['long'] = fuzz.trimf(scen.universe,[1.4,2,2])
scen.view()
```

```python
sright['short'] = fuzz.trimf(sright.universe,[0,0,0.8])
sright['medium'] = fuzz.trimf(sright.universe,[0.2,1,1.7])
sright['long'] = fuzz.trimf(sright.universe,[1.4,2,2])
sright.view()
```

```python
mleft['negative'] = fuzz.trimf(mleft.universe,[-100,-100,0])
mleft['zero'] = fuzz.trimf(mleft.universe,[-100,0,100])
mleft['positive'] = fuzz.trimf(mleft.universe,[0,100,100])
mleft.view()
```

```python
mright['negative'] = fuzz.trimf(mright.universe,[-100,-100,0])
mright['zero'] = fuzz.trimf(mright.universe,[-100,0,100])
mright['positive'] = fuzz.trimf(mright.universe,[0,100,100])
mright.view()
```

```python
rule1 = ctrl.Rule(sleft['short'] & scen['short'] & sright['short'], 
                       (mleft['negative'], mright['negative']))
rule2 = ctrl.Rule(sleft['short'] & scen['short'] & sright['medium'], 
                       (mleft['negative'], mright['negative']))
rule3 = ctrl.Rule(sleft['short'] & scen['short'] & sright['long'], 
                       (mleft['positive'], mright['zero']))
rule4 = ctrl.Rule(sleft['short'] & scen['medium'] & sright['short'], 
                       (mleft['positive'], mright['positive']))
rule5 = ctrl.Rule(sleft['short'] & scen['medium'] & sright['medium'], 
                       (mleft['positive'], mright['zero']))
rule6 = ctrl.Rule(sleft['short'] & scen['medium'] & sright['long'], 
                       (mleft['positive'], mright['zero']))
rule7 = ctrl.Rule(sleft['short'] & scen['long'] & sright['short'], 
                       (mleft['positive'], mright['positive']))
rule8 = ctrl.Rule(sleft['short'] & scen['long'] & sright['medium'], 
                       (mleft['positive'], mright['zero']))
rule9 = ctrl.Rule(sleft['short'] & scen['long'] & sright['long'], 
                       (mleft['positive'], mright['zero']))
rule10 = ctrl.Rule(sleft['medium'] & scen['short'] & sright['short'], 
                       (mleft['zero'], mright['positive']))
rule11 = ctrl.Rule(sleft['medium'] & scen['short'] & sright['medium'], 
                       (mleft['negative'], mright['positive']))
rule12 = ctrl.Rule(sleft['medium'] & scen['short'] & sright['long'], 
                       (mleft['positive'], mright['zero']))
rule13 = ctrl.Rule(sleft['medium'] & scen['medium'] & sright['short'], 
                       (mleft['zero'], mright['positive']))
rule14 = ctrl.Rule(sleft['medium'] & scen['medium'] & sright['medium'], 
                       (mleft['positive'], mright['positive']))
rule15 = ctrl.Rule(sleft['medium'] & scen['medium'] & sright['long'], 
                       (mleft['positive'], mright['positive']))
rule16 = ctrl.Rule(sleft['medium'] & scen['long'] & sright['short'], 
                       (mleft['zero'], mright['positive']))
rule17 = ctrl.Rule(sleft['medium'] & scen['long'] & sright['medium'], 
                       (mleft['positive'], mright['positive']))
rule18 = ctrl.Rule(sleft['medium'] & scen['long'] & sright['long'], 
                       (mleft['positive'], mright['positive']))
rule19 = ctrl.Rule(sleft['long'] & scen['short'] & sright['short'], 
                       (mleft['zero'], mright['positive']))
rule20 = ctrl.Rule(sleft['long'] & scen['short'] & sright['medium'], 
                       (mleft['zero'], mright['positive']))
rule21 = ctrl.Rule(sleft['long'] & scen['short'] & sright['long'], 
                       (mleft['negative'], mright['negative']))
rule22 = ctrl.Rule(sleft['long'] & scen['medium'] & sright['short'], 
                       (mleft['zero'], mright['positive']))
rule23 = ctrl.Rule(sleft['long'] & scen['medium'] & sright['medium'], 
                       (mleft['zero'], mright['positive']))
rule24 = ctrl.Rule(sleft['long'] & scen['medium'] & sright['long'], 
                       (mleft['positive'], mright['positive']))
rule25 = ctrl.Rule(sleft['long'] & scen['long'] & sright['short'], 
                       (mleft['zero'], mright['positive']))
rule26 = ctrl.Rule(sleft['long'] & scen['long'] & sright['medium'], 
                       (mleft['positive'], mright['positive']))
rule27 = ctrl.Rule(sleft['long'] & scen['long'] & sright['long'], 
                       (mleft['positive'], mright['positive']))
```

```python
wheel_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27])
wheel_system=ctrl.ControlSystemSimulation(wheel_ctrl)
```

```python
wheel_system.input['sright']=float(input('Distance from obstacle to right sensor: '))
wheel_system.input['scen']=float(input('Distance from obstacle to center sensor: '))
wheel_system.input['sleft']=float(input('Distance from obstacle to left sensor: '))
wheel_system.compute()
print('Speed of right wheel: ',int(wheel_system.output['mright']),'rpm')
print('Speed of left wheel: ',int(wheel_system.output['mleft']),'rpm')
mright.view(sim=wheel_system)
mleft.view(sim=wheel_system)
```
