```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
weight = ctrl.Antecedent(np.arange(100, 6001, 1), 'weight') 
water = ctrl.Antecedent(np.arange(0, 81, 1), 'water')
material = ctrl.Antecedent(np.arange(0, 101, 0.001), 'material')
trans = ctrl.Antecedent(np.arange(0, 101, 0.01), 'trans')
time = ctrl.Consequent(np.arange(10, 91, 1), 'time')
```

```python
weight['light'] = fuzz.trimf(weight.universe,[100,100,2500])
weight['normal'] = fuzz.trimf(weight.universe,[800,2700,5500])
weight['heavy'] = fuzz.trimf(weight.universe,[3300,6000,6000])
weight.view()
```

```python
water['cold'] = fuzz.trimf(water.universe,[0,0,30])
water['warm'] = fuzz.trimf(water.universe,[20,40,65])
water['hot'] = fuzz.trimf(water.universe,[60,80,80])
water.view()
```

```python
material['soft'] = fuzz.trapmf(material.universe,[0,0,30,30])
material['normal'] = fuzz.trapmf(material.universe,[30,30,60,60])
material['hard'] = fuzz.trapmf(material.universe,[60,60,100,100])
material.view()
```

```python
trans['low'] = fuzz.trimf(trans.universe,[0,0,30])
trans['medium'] = fuzz.trimf(trans.universe,[20,50,70])
trans['high'] = fuzz.trimf(trans.universe,[65,100,100])
trans.view()
```

```python
time['short'] = fuzz.trimf(time.universe,[10,10,20])
time['medium'] = fuzz.trimf(time.universe,[15,45,65])
time['long'] = fuzz.trimf(time.universe,[60,90,90])
time.view()
```
